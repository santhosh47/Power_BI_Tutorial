Whether your data is in the form of an Excel spreadsheet or if it's on cloud, you can easily connect
your data from different data sources, visualize and analyze your data, and share it with anyone or everyone
with the help of Microsoft Power BI. 

Power BI has three parts - **Power BI Desktop, Power BI Service, and Power BI Mobile Apps.**

#### **Power BI Desktop:** 
Power BI Desktop, which is a free Windows desktop application,
is used to create reports, which is then published
to Power BI Service. 

#### **Power BI Service:**
Power BI Service is an online service which helps you view reports and dashboards.
